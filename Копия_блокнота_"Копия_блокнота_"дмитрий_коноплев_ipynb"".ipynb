```python
from ultralytics import YOLO

# Load a pre-trained model
model = YOLO("yolov8n.pt")
```

```python
# –£—Å—Ç–∞–Ω–æ–≤–∫–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫
!pip install ultralytics

# –ò–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import numpy as np

# –ó–∞–≥—Ä—É–∑–∫–∞ –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ YOLOv8
model = YOLO("yolov8n.pt")  # –í—ã –º–æ–∂–µ—Ç–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥—Ä—É–≥–∏–µ –≤–µ—Ä—Å–∏–∏, –Ω–∞–ø—Ä–∏–º–µ—Ä yolov8s.pt

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –∏ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
def predict_image(image_path):
    # –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
    results = model.predict(image_path)

    # –°—á–µ—Ç—á–∏–∫–∏ –¥–ª—è –º–µ—Ç—Ä–∏–∫
    true_positives = 0
    false_positives = 0
    false_negatives = 0

    # –ß—Ç–µ–Ω–∏–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    img = cv2.imread(image_path)

    # –û–±—Ä–∞–±–æ—Ç–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    for result in results:
        boxes = result.boxes  # –ü–æ–ª—É—á–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –±–æ–∫—Å—ã
        for box in boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # –ö–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –±–æ–∫—Å–∞
            class_id = int(box.cls)  # –ö–ª–∞—Å—Å –æ–±—ä–µ–∫—Ç–∞
            confidence = box.conf  # –£–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è

            # –†–∏—Å—É–µ–º –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ –≤–æ–∫—Ä—É–≥ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–Ω–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
            label = f"Class: {model.names[class_id]}, Confidence: {confidence:.2f}"
            cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

            # –£—Å–ª–æ–≤–∏–µ –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ TP/FP/FN (–ø—Ä–∏–º–µ—Ä–Ω–∞—è –ª–æ–≥–∏–∫–∞)
            if confidence > 0.5:  # –ü—Ä–∏–º–µ—Ä: –µ—Å–ª–∏ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –±–æ–ª—å—à–µ 0.5 —Å—á–∏—Ç–∞–µ—Ç—Å—è TP
                true_positives += 1
            else:
                false_positives += 1

    # –î–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏ –¥–æ–±–∞–≤–∏–º —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è –ª–æ–∂–Ω–æ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö (–º–æ–∂–Ω–æ –∑–∞–º–µ–Ω–∏—Ç—å –Ω–∞ —Ä–µ–∞–ª—å–Ω–æ–µ)
    false_negatives = max(0, 10 - true_positives)  # –ü—Ä–∏–º–µ—Ä: –≤—Å–µ–≥–æ –æ–∂–∏–¥–∞–µ—Ç—Å—è 10 –æ–±—ä–µ–∫—Ç–æ–≤

    # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è —Å –∞–Ω–Ω–æ—Ç–∞—Ü–∏—è–º–∏
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    return true_positives, false_positives, false_negatives

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Ç–æ—á–Ω–æ—Å—Ç–∏ –∏ –æ—à–∏–±–æ–∫
def plot_metrics(tp, fp, fn):
    total_predictions = tp + fp + fn

    if total_predictions == 0:
        accuracy = 0
        error_rate = 0
    else:
        accuracy = tp / total_predictions * 100
        error_rate = (fp + fn) / total_predictions * 100

    metrics = ['True Positives', 'False Positives', 'False Negatives']
    values = [tp, fp, fn]

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.bar(metrics, values, color=['green', 'red', 'orange'])
    plt.title('Predictions Count')
    plt.ylabel('Count')

    plt.subplot(1, 2, 2)
    plt.bar(['Accuracy', 'Error Rate'], [accuracy, error_rate], color=['blue', 'purple'])
    plt.title('Model Performance')
    plt.ylabel('Percentage (%)')

    plt.tight_layout()
    plt.show()

# –ü—É—Ç—å –∫ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—é –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ (–∑–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ —Å–≤–æ–π –ø—É—Ç—å)
image_path = "/content/drive/MyDrive/proplex_1/inclusions_all/dataset/images/test/179469f5-–∫–∞–¥—Ä_0335.png"  # –ó–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ –ø—É—Ç—å –∫ –≤–∞—à–µ–º—É –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—é

# –í—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –∏ –ø–æ–ª—É—á–µ–Ω–∏—è –º–µ—Ç—Ä–∏–∫
true_positives, false_positives, false_negatives = predict_image(image_path)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Ç–æ—á–Ω–æ—Å—Ç–∏ –∏ –æ—à–∏–±–æ–∫
plot_metrics(true_positives, false_positives, false_negatives)
```

```python
# –£—Å—Ç–∞–Ω–æ–≤–∫–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫
import ultralytics

# –ò–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import numpy as np

# –ó–∞–≥—Ä—É–∑–∫–∞ –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ YOLOv8
model = YOLO("yolov8n.pt")  # –í—ã –º–æ–∂–µ—Ç–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥—Ä—É–≥–∏–µ –≤–µ—Ä—Å–∏–∏, –Ω–∞–ø—Ä–∏–º–µ—Ä yolov8s.pt
image_path = "/content/drive/MyDrive/proplex_1/inclusions_all/dataset/images/test/07de3c02-frame373.jpg"

#image_path = "/content/drive/MyDrive/proplex_1/inclusions_all/dataset/images/test/179469f5-–∫–∞–¥—Ä_0335.png"
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –∏ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
def predict_image(image_path):
    # –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
    results = model.predict(image_path)

    # –°—á–µ—Ç—á–∏–∫–∏ –¥–ª—è –º–µ—Ç—Ä–∏–∫
    true_positives = 0
    false_positives = 0
    false_negatives = 0

    # –ß—Ç–µ–Ω–∏–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    img = cv2.imread(image_path)

    # –û–±—Ä–∞–±–æ—Ç–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    for result in results:
        boxes = result.boxes  # –ü–æ–ª—É—á–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –±–æ–∫—Å—ã
        for box in boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # –ö–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –±–æ–∫—Å–∞
            class_id = int(box.cls)  # –ö–ª–∞—Å—Å –æ–±—ä–µ–∫—Ç–∞
            confidence = box.conf  # –£–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è

            # –†–∏—Å—É–µ–º –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ –≤–æ–∫—Ä—É–≥ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–Ω–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
            label = f"Class: {model.names[class_id]}, Confidence: {confidence:.2f}"
            cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

            # –£—Å–ª–æ–≤–∏–µ –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ TP/FP/FN (–ø—Ä–∏–º–µ—Ä–Ω–∞—è –ª–æ–≥–∏–∫–∞)
            if confidence > 0.5:  # –ü—Ä–∏–º–µ—Ä: –µ—Å–ª–∏ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –±–æ–ª—å—à–µ 0.5 —Å—á–∏—Ç–∞–µ—Ç—Å—è TP
                true_positives += 1
            else:
                false_positives += 1

    # –î–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏ –¥–æ–±–∞–≤–∏–º —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è –ª–æ–∂–Ω–æ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö (–º–æ–∂–Ω–æ –∑–∞–º–µ–Ω–∏—Ç—å –Ω–∞ —Ä–µ–∞–ª—å–Ω–æ–µ)
    false_negatives = max(0, 10 - true_positives)  # –ü—Ä–∏–º–µ—Ä: –≤—Å–µ–≥–æ –æ–∂–∏–¥–∞–µ—Ç—Å—è 10 –æ–±—ä–µ–∫—Ç–æ–≤

    # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è —Å –∞–Ω–Ω–æ—Ç–∞—Ü–∏—è–º–∏
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    return true_positives, false_positives, false_negatives

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Ç–æ—á–Ω–æ—Å—Ç–∏ –∏ –æ—à–∏–±–æ–∫
def plot_metrics(tp, fp, fn):
    total_predictions = tp + fp + fn

    if total_predictions == 0:
        accuracy = 0
        error_rate = 0
    else:
        accuracy = tp / total_predictions * 100
        error_rate = (fp + fn) / total_predictions * 100

    metrics = ['True Positives', 'False Positives', 'False Negatives']
    values = [tp, fp, fn]

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.bar(metrics, values, color=['green', 'red', 'orange'])
    plt.title('Predictions Count')
    plt.ylabel('Count')

    plt.subplot(1, 2, 2)
    plt.bar(['Accuracy', 'Error Rate'], [accuracy, error_rate], color=['blue', 'purple'])
    plt.title('Model Performance')
    plt.ylabel('Percentage (%)')

    plt.tight_layout()
    plt.show()

# –ü—É—Ç—å –∫ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—é –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ (–∑–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ —Å–≤–æ–π –ø—É—Ç—å)
image_path = "/content/drive/MyDrive/proplex_1/inclusions_all/dataset/images/test/02562471-frame429.jpg"  # –ó–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ –ø—É—Ç—å –∫ –≤–∞—à–µ–º—É –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—é

# –í—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –∏ –ø–æ–ª—É—á–µ–Ω–∏—è –º–µ—Ç—Ä–∏–∫
#true_positives, false_positives, false_negatives = predict_image(image_path)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Ç–æ—á–Ω–æ—Å—Ç–∏ –∏ –æ—à–∏–±–æ–∫
plot_metrics(true_positives, false_positives, false_negatives)
```

```python


# –ó–∞–≥—Ä—É–∑–∫–∞ –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ YOLOv8
model = YOLO("yolov8n.pt")  # –ú–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥—Ä—É–≥–∏–µ –≤–µ—Ä—Å–∏–∏, –Ω–∞–ø—Ä–∏–º–µ—Ä yolov8s.pt

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –¥–µ—Ñ–µ–∫—Ç–æ–≤ –æ–∫–æ–Ω–Ω–æ–≥–æ –ø—Ä–æ—Ñ–∏–ª—è
def process_window_defects(image_path):
    # –ß—Ç–µ–Ω–∏–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
    image = cv2.imread(image_path)

    # –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
    results = model.predict(image_path)

    # –°–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    true_positives = 0
    false_positives = 0
    false_negatives = 0

    # –û–±—Ä–∞–±–æ—Ç–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    for result in results:
        boxes = result.boxes  # –ü–æ–ª—É—á–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –±–æ–∫—Å—ã
        for box in boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # –ö–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –±–æ–∫—Å–∞
            class_id = int(box.cls)  # –ö–ª–∞—Å—Å –æ–±—ä–µ–∫—Ç–∞
            confidence = box.conf  # –£–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è

            # –†–∏—Å—É–µ–º –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ –≤–æ–∫—Ä—É–≥ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–Ω–æ–≥–æ –¥–µ—Ñ–µ–∫—Ç–∞
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
            label = f"Class: {model.names[class_id]}, Confidence: {confidence:.2f}"
            cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

            # –ó–¥–µ—Å—å –º–æ–∂–Ω–æ –¥–æ–±–∞–≤–∏—Ç—å –ª–æ–≥–∏–∫—É –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ TP/FP/FN
            if confidence > 0.5:  # –ü—Ä–∏–º–µ—Ä: –µ—Å–ª–∏ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –±–æ–ª—å—à–µ 0.5 —Å—á–∏—Ç–∞–µ—Ç—Å—è TP
                true_positives += 1
            else:
                false_positives += 1

    # –î–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏ –¥–æ–±–∞–≤–∏–º —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è –ª–æ–∂–Ω–æ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö (–º–æ–∂–Ω–æ –∑–∞–º–µ–Ω–∏—Ç—å –Ω–∞ —Ä–µ–∞–ª—å–Ω–æ–µ)
    false_negatives = max(0, 10 - true_positives)  # –ü—Ä–∏–º–µ—Ä: –≤—Å–µ–≥–æ –æ–∂–∏–¥–∞–µ—Ç—Å—è 10 –æ–±—ä–µ–∫—Ç–æ–≤

    # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
    cv2.imshow("Detected Defects", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    return true_positives, false_positives, false_negatives

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Ç–æ—á–Ω–æ—Å—Ç–∏ –∏ –æ—à–∏–±–æ–∫
def plot_metrics(tp, fp, fn):
    total_predictions = tp + fp + fn

    if total_predictions == 0:
        accuracy = 0
        error_rate = 0
    else:
        accuracy = tp / total_predictions * 100
        error_rate = (fp + fn) / total_predictions * 100

    metrics = ['True Positives', 'False Positives', 'False Negatives']
    values = [tp, fp, fn]

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.bar(metrics, values, color=['green', 'red', 'orange'])
    plt.title('Predictions Count')
    plt.ylabel('Count')

    plt.subplot(1, 2, 2)
    plt.bar(['Accuracy', 'Error Rate'], [accuracy, error_rate], color=['blue', 'purple'])
    plt.title('Model Performance')
    plt.ylabel('Percentage (%)')

    plt.tight_layout()
    plt.show()

# –ü—É—Ç—å –∫ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—é –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏
image_path = "path/to/your/window_defect_image.jpg"  # –ó–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ –ø—É—Ç—å –∫ –≤–∞—à–µ–º—É –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—é

# –í—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –∏ –ø–æ–ª—É—á–µ–Ω–∏—è –º–µ—Ç—Ä–∏–∫
true_positives, false_positives, false_negatives = process_window_defects(image_path)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Ç–æ—á–Ω–æ—Å—Ç–∏ –∏ –æ—à–∏–±–æ–∫
plot_metrics(true_positives, false_positives, false_negatives)
```

```python
# –ò–º–ø–æ—Ä—Ç—ã –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

# –ó–∞–≥—Ä—É–∑–∫–∞ –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ YOLOv8
model = YOLO("yolov8n.pt")  # –ú–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥—Ä—É–≥–∏–µ –≤–µ—Ä—Å–∏–∏, –Ω–∞–ø—Ä–∏–º–µ—Ä yolov8s.pt

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –¥–µ—Ñ–µ–∫—Ç–æ–≤ –æ–∫–æ–Ω–Ω–æ–≥–æ –ø—Ä–æ—Ñ–∏–ª—è
def process_window_defects(image_path):
    # –ß—Ç–µ–Ω–∏–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
    image = cv2.imread(image_path)
    if image is None:
        print(f"–û—à–∏–±–∫–∞: –ù–µ —É–¥–∞–ª–æ—Å—å –∑–∞–≥—Ä—É–∑–∏—Ç—å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –ø–æ –ø—É—Ç–∏ {image_path}")
        return None, None, None

    # –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
    results = model.predict(image_path)

    # –°–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    true_positives = 0
    false_positives = 0
    false_negatives = 0

    # –û–±—Ä–∞–±–æ—Ç–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
    for result in results:
        for box in result.boxes:
            # –ü–æ–ª—É—á–∞–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –±–æ–∫—Å–∞, –∫–ª–∞—Å—Å –æ–±—ä–µ–∫—Ç–∞ –∏ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # –ö–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –±–æ–∫—Å–∞ (–ª–µ–≤–∞—è –≤–µ—Ä—Ö–Ω—è—è –∏ –ø—Ä–∞–≤–∞—è –Ω–∏–∂–Ω—è—è —Ç–æ—á–∫–∏)
            class_id = int(box.cls)  # –ö–ª–∞—Å—Å –æ–±—ä–µ–∫—Ç–∞
            confidence = box.conf.item()  # –£–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è

            # –†–∏—Å—É–µ–º –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ –≤–æ–∫—Ä—É–≥ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–Ω–æ–≥–æ –¥–µ—Ñ–µ–∫—Ç–∞
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
            label = f"Class: {model.names[class_id]}, Confidence: {confidence:.2f}"
            cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

            # –õ–æ–≥–∏–∫–∞ –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ TP/FP/FN
            if confidence > 0.5:  # –ü—Ä–∏–º–µ—Ä: –µ—Å–ª–∏ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –±–æ–ª—å—à–µ 0.5, —Å—á–∏—Ç–∞–µ—Ç—Å—è TP
                true_positives += 1
            else:
                false_positives += 1

    # –õ–æ–∂–Ω–æ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ (–º–æ–∂–Ω–æ –∑–∞–º–µ–Ω–∏—Ç—å –Ω–∞ —Ä–µ–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, –µ—Å–ª–∏ –∏–∑–≤–µ—Å—Ç–Ω–æ –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–µ—Ñ–µ–∫—Ç–æ–≤)
    false_negatives = max(0, 10 - true_positives)  # –ü—Ä–∏–º–µ—Ä: –≤—Å–µ–≥–æ –æ–∂–∏–¥–∞–µ—Ç—Å—è 10 –æ–±—ä–µ–∫—Ç–æ–≤

    # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
    cv2.imshow("Detected Defects", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    return true_positives, false_positives, false_negatives

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Ç–æ—á–Ω–æ—Å—Ç–∏ –∏ –æ—à–∏–±–æ–∫
def plot_metrics(tp, fp, fn):
    if tp is None or fp is None or fn is None:
        print("–û—à–∏–±–∫–∞: –ú–µ—Ç—Ä–∏–∫–∏ –Ω–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω—ã.")
        return

    total_predictions = tp + fp + fn

    if total_predictions == 0:
        accuracy = 0
        error_rate = 0
    else:
        accuracy = tp / total_predictions * 100
        error_rate = (fp + fn) / total_predictions * 100

    metrics = ['True Positives', 'False Positives', 'False Negatives']
    values = [tp, fp, fn]

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.bar(metrics, values, color=['green', 'red', 'orange'])
    plt.title('Predictions Count')
    plt.ylabel('Count')

    plt.subplot(1, 2, 2)
    plt.bar(['Accuracy', 'Error Rate'], [accuracy, error_rate], color=['blue', 'purple'])
    plt.title('Model Performance')
    plt.ylabel('Percentage (%)')

    plt.tight_layout()
    plt.show()

# –ü—É—Ç—å –∫ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—é –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏
#image_path = "/content/drive/MyDrive/proplex_1/inclusions_all/image1.jpg"  # –£–∫–∞–∂–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–µ –∏–º—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è



# –í—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –∏ –ø–æ–ª—É—á–µ–Ω–∏—è –º–µ—Ç—Ä–∏–∫
true_positives, false_positives, false_negatives = process_window_defects(image_path)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Ç–æ—á–Ω–æ—Å—Ç–∏ –∏ –æ—à–∏–±–æ–∫
plot_metrics(true_positives, false_positives, false_negatives)
```

```python
import os

# –ü—É—Ç—å –∫ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º–∏
image_directory = "/content/drive/MyDrive/proplex_1/inclusions_all"

# –ü–æ–ª—É—á–µ–Ω–∏–µ —Å–ø–∏—Å–∫–∞ –≤—Å–µ—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –≤ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏
image_files = [os.path.join(image_directory, f) for f in os.listdir(image_directory) if f.endswith(".jpg")]

# –û–±—Ä–∞–±–æ—Ç–∫–∞ –≤—Å–µ—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –≤ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏
for image_path in image_files:
    print(f"–û–±—Ä–∞–±–æ—Ç–∫–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è: {image_path}")
    true_positives, false_positives, false_negatives = process_window_defects(image_path)
    plot_metrics(true_positives, false_positives, false_negatives)
```

```python
# –ò–º–ø–æ—Ä—Ç—ã –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

# –ó–∞–≥—Ä—É–∑–∫–∞ –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ YOLOv8
model = YOLO("yolov8n.pt")  # –ú–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥—Ä—É–≥–∏–µ –≤–µ—Ä—Å–∏–∏, –Ω–∞–ø—Ä–∏–º–µ—Ä yolov8s.pt

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –¥–µ—Ñ–µ–∫—Ç–æ–≤ –æ–∫–æ–Ω–Ω–æ–≥–æ –ø—Ä–æ—Ñ–∏–ª—è
def process_window_defects(image_path):
    # –ß—Ç–µ–Ω–∏–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
    image = cv2.imread(image_path)

    # –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
    results = model.predict(image_path)

    # –°–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    true_positives = 0
    false_positives = 0
    false_negatives = 0

    # –û–±—Ä–∞–±–æ—Ç–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    for result in results:
        boxes = result.boxes  # –ü–æ–ª—É—á–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –±–æ–∫—Å—ã
        for box in boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # –ö–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –±–æ–∫—Å–∞
            class_id = int(box.cls)  # –ö–ª–∞—Å—Å –æ–±—ä–µ–∫—Ç–∞
            confidence = box.conf  # –£–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è

            # –†–∏—Å—É–µ–º –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ –≤–æ–∫—Ä—É–≥ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–Ω–æ–≥–æ –¥–µ—Ñ–µ–∫—Ç–∞
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
            label = f"Class: {model.names[class_id]}, Confidence: {confidence:.2f}"
            cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

            # –ó–¥–µ—Å—å –º–æ–∂–Ω–æ –¥–æ–±–∞–≤–∏—Ç—å –ª–æ–≥–∏–∫—É –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ TP/FP/FN
            if confidence > 0.5:  # –ü—Ä–∏–º–µ—Ä: –µ—Å–ª–∏ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –±–æ–ª—å—à–µ 0.5, —Å—á–∏—Ç–∞–µ—Ç—Å—è TP
                true_positives += 1
            else:
                false_positives += 1

    # –î–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏ –¥–æ–±–∞–≤–∏–º —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è –ª–æ–∂–Ω–æ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö (–º–æ–∂–Ω–æ –∑–∞–º–µ–Ω–∏—Ç—å –Ω–∞ —Ä–µ–∞–ª—å–Ω–æ–µ)
    false_negatives = max(0, 10 - true_positives)  # –ü—Ä–∏–º–µ—Ä: –≤—Å–µ–≥–æ –æ–∂–∏–¥–∞–µ—Ç—Å—è 10 –æ–±—ä–µ–∫—Ç–æ–≤

    # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
    cv2.imshow("Detected Defects", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    return true_positives, false_positives, false_negatives

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Ç–æ—á–Ω–æ—Å—Ç–∏ –∏ –æ—à–∏–±–æ–∫
def plot_metrics(tp, fp, fn):
    total_predictions = tp + fp + fn

    if total_predictions == 0:
        accuracy = 0
        error_rate = 0
    else:
        accuracy = tp / total_predictions * 100
        error_rate = (fp + fn) / total_predictions * 100

    metrics = ['True Positives', 'False Positives', 'False Negatives']
    values = [tp, fp, fn]

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.bar(metrics, values, color=['green', 'red', 'orange'])
    plt.title('Predictions Count')
    plt.ylabel('Count')

    plt.subplot(1, 2, 2)
    plt.bar(['Accuracy', 'Error Rate'], [accuracy, error_rate], color=['blue', 'purple'])
    plt.title('Model Performance')
    plt.ylabel('Percentage (%)')

    plt.tight_layout()
    plt.show()

# –ü—É—Ç—å –∫ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—é –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏
image_path = "path/to/your/window_defect_image.jpg"  # –ó–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ –ø—É—Ç—å –∫ –≤–∞—à–µ–º—É –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—é

# –í—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –∏ –ø–æ–ª—É—á–µ–Ω–∏—è –º–µ—Ç—Ä–∏–∫
true_positives, false_positives, false_negatives = process_window_defects(image_path)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Ç–æ—á–Ω–æ—Å—Ç–∏ –∏ –æ—à–∏–±–æ–∫
plot_metrics(true_positives, false_positives, false_negatives)
```
